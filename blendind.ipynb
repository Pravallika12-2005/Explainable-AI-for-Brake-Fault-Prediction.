```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import sklearn
import seaborn as sns
from sklearn.experimental import enable_iterative_imputer
from sklearn.impute import IterativeImputer
from collections import Counter
from sklearn.ensemble import RandomForestClassifier
from sklearn.feature_selection import RFE
from sklearn.metrics import confusion_matrix, accuracy_score, classification_report
import joblib  # Import joblib for saving the model

# Load the datasets
train_df = pd.read_csv('C:/Users/system -2/Desktop/Break_fault_in_Heavy_Transports/Dataset/aps_failure_training_set.csv', na_values=["na"])
test_df = pd.read_csv('C:/Users/system -2/Desktop/Break_fault_in_Heavy_Transports/Dataset/aps_failure_test_set.csv', na_values=["na"])

# Combine the datasets
df = pd.concat([train_df, test_df], ignore_index=True)

# Dropping columns with more than 50% missing values
drop_cols = []
for i in df.columns:
    null_percentage = df[i].isnull().sum() / len(df) * 100
    if null_percentage > 50:
        print(i)
        drop_cols.append(str(i))

df.drop(drop_cols, axis=1, inplace=True)

# Replace categorical values with integers and infer object types
df['class'] = df['class'].replace(['neg', 'pos'], [0, 1]).infer_objects()

# Handle missing values
missing_vals = {}
for i in df.columns:
    null_percentage = df[i].isnull().sum() / len(df) * 100
    if null_percentage > 0:
        missing_vals[i] = null_percentage

missing_vals = sorted(missing_vals.items(), key=lambda x: x[1], reverse=True)

na_5 = [items[0] for items in missing_vals if items[1] < 5]
df.dropna(subset=na_5, inplace=True)

na_5to15 = [items[0] for items in missing_vals if items[1] >= 5 and items[1] < 15]
df[na_5to15] = df[na_5to15].fillna(df[na_5to15].median())

na_50 = [items[0] for items in missing_vals if items[1] >= 15 and items[1] < 50]
imputer = IterativeImputer(n_nearest_features=10, max_iter=10)
df_filled = pd.DataFrame(imputer.fit_transform(df[na_50]), columns=na_50)
df[na_50] = df_filled

Train = df[:60000]
Test = df[60000:]

y_train = Train['class'].copy()
X_train = Train.drop(['class'], axis=1).copy()
y_test = Test['class'].copy()
X_test = Test.drop(['class'], axis=1).copy()

def Identifier_Bin_count(x):
    prefix = []
    for name in x.columns:
        prefix.append(name.split('_')[0])
    counter = Counter(prefix)
    return list(counter.keys()), list(counter.values())

feature_prefix, bin_count = Identifier_Bin_count(X_train)
hist_identifier = [feature_prefix[i] for i, j in enumerate(bin_count) if j == 10]
hist_features = [i for i in X_train.columns if i.split('_')[0] in hist_identifier]

def get_top(df, y, n):
    feature_select = RFE(estimator=RandomForestClassifier(n_estimators=150, max_depth=5, random_state=1),
                         n_features_to_select=n,
                         verbose=5)
    feature_select.fit(df, y)
    top_f = [c for i, c in enumerate(df.columns.tolist()) if feature_select.support_[i]]
    return top_f

top_feature_hist = get_top(X_train[hist_features], y_train, 15)
print("The top features selected after Recursive Feature Elimination are: \n", top_feature_hist)

model = RandomForestClassifier()
model.fit(X_train[top_feature_hist], y_train)
y_pred = model.predict(X_test[top_feature_hist])

# Calculate metrics
accuracy = accuracy_score(y_test, y_pred)
report = classification_report(y_test, y_pred, target_names=['Negative', 'Positive'], output_dict=True)

# Print results
print("Accuracy: {:.2f}%".format(accuracy * 100))
print("Recall (Positive): {:.2f}%".format(report['Positive']['recall'] * 100))
print("Precision (Positive): {:.2f}%".format(report['Positive']['precision'] * 100))
print("F1 Score (Positive): {:.2f}".format(report['Positive']['f1-score']))

# Save the model to a .pkl file
joblib.dump(model, 'C:/Users/system -2/Desktop/Break_fault_in_Heavy_Transports/Dataset/random_forest2_model.pkl')
print("Model saved successfully!")
```
